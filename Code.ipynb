```python
# ============================
# 1. Import Necessary Libraries
# ============================

# Install torch and torchvision if not already installed
# (Uncomment the following line if running in an environment where torch isn't installed)
# !pip install torch torchvision --quiet

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.utils.prune as prune
import time
import pandas as pd

# ============================
# 2. Set Device to CPU or CUDA
# ============================

# Set device to CUDA if available, else CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ============================
# 3. Define Data Transformations and Load Datasets
# ============================

# Define transformations: convert images to tensors and normalize them
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # MNIST mean and std
])

# Define batch size
batch_size = 64

# Download and load training and test datasets
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Define DataLoaders with drop_last=True to ensure all batches are the same size
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, drop_last=True)

print(f"Number of training batches: {len(train_loader)}")
print(f"Number of testing batches: {len(test_loader)}")

# ============================
# 4. Display a Single Training Example
# ============================

# Function to unnormalize and convert tensor to numpy array for display
def imshow(img, title):
    img = img.numpy().transpose((1, 2, 0))  # Convert from Tensor image
    mean = np.array([0.1307])
    std = np.array([0.3081])
    img = std * img + mean  # Unnormalize
    img = np.clip(img, 0, 1)  # Clip to valid range
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Get one batch of training data
dataiter = iter(train_loader)
images, labels = next(dataiter)

# Select the first image and label from the batch
single_img = images[0]
single_label = labels[0]

# Display the single training example
print("Displaying a single training example:")
imshow(single_img.cpu(), f"Label: {single_label.item()}")

# ============================
# 5. Define the Expanded CNN Model
# ============================

class ExpandedCNN(nn.Module):
    def __init__(self, conv1_out_channels=32, conv2_out_channels=64, conv3_out_channels=128, conv4_out_channels=256):
        super(ExpandedCNN, self).__init__()
        # First block
        self.conv1 = nn.Conv2d(1, conv1_out_channels, kernel_size=3, padding=1)  # 1x28x28 -> conv1_out_channels x28x28
        self.relu1 = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(conv1_out_channels, conv2_out_channels, kernel_size=3, padding=1)  # conv1_out_channels x28x28 -> conv2_out_channels x28x28
        self.relu2 = nn.ReLU(inplace=True)
        self.pool1 = nn.MaxPool2d(2, 2)  # conv2_out_channels x28x28 -> conv2_out_channels x14x14
        self.dropout1 = nn.Dropout(0.25)

        # Second block
        self.conv3 = nn.Conv2d(conv2_out_channels, conv3_out_channels, kernel_size=3, padding=1)  # conv2_out_channels x14x14 -> conv3_out_channels x14x14
        self.relu3 = nn.ReLU(inplace=True)
        self.conv4 = nn.Conv2d(conv3_out_channels, conv4_out_channels, kernel_size=3, padding=1)  # conv3_out_channels x14x14 -> conv4_out_channels x14x14
        self.relu4 = nn.ReLU(inplace=True)
        self.pool2 = nn.MaxPool2d(2, 2)  # conv4_out_channels x14x14 -> conv4_out_channels x7x7
        self.dropout2 = nn.Dropout(0.25)

        # Fully connected layers
        self.fc1 = nn.Linear(conv4_out_channels * 7 * 7, 512)
        self.relu5 = nn.ReLU(inplace=True)
        self.dropout3 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 10)  # 10 output classes for MNIST

    def forward(self, x):
        # First block
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool1(x)
        x = self.dropout1(x)

        # Second block
        x = self.conv3(x)
        x = self.relu3(x)
        x = self.conv4(x)
        x = self.relu4(x)
        x = self.pool2(x)
        x = self.dropout2(x)

        # Fully connected layers
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc1(x)
        x = self.relu5(x)
        x = self.dropout3(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)  # Log-Softmax for classification

# ============================
# 6. Initialize and Train the Original Model
# ============================

def train_and_evaluate(model, train_loader, test_loader, device, epochs=10):  # Changed to 10 epochs
    """
    Trains the model for a specified number of epochs and evaluates on the test set.

    Args:
        model (nn.Module): The neural network model.
        train_loader (DataLoader): DataLoader for the training dataset.
        test_loader (DataLoader): DataLoader for the test dataset.
        device (torch.device): Device to perform computations on.
        epochs (int): Number of training epochs.

    Returns:
        test_accuracy (float): Accuracy of the model on the test set after training.
        training_time (float): Time taken to train the model in seconds.
    """
    criterion = nn.NLLLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    start_time = time.time()

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()  # Zero the parameter gradients
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = correct / total

        # Evaluate on test data
        model.eval()
        test_loss = 0.0
        correct_test = 0
        total_test = 0

        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                test_loss += loss.item() * images.size(0)
                _, predicted = torch.max(outputs.data, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        epoch_test_loss = test_loss / len(test_loader.dataset)
        epoch_test_acc = correct_test / total_test

        print(f"Epoch {epoch}/{epochs} | "
              f"Train Loss: {epoch_loss:.4f} | Train Acc: {epoch_acc:.4f} | "
              f"Test Loss: {epoch_test_loss:.4f} | Test Acc: {epoch_test_acc:.4f}")

    end_time = time.time()
    training_time = end_time - start_time

    return epoch_test_acc, training_time

# Initialize the original model
original_model = ExpandedCNN().to(device)
print("\nOriginal Model Architecture:")
print(original_model)

# Train and evaluate the original model
print("\nTraining Original Model:")
original_test_acc, original_train_time = train_and_evaluate(
    original_model, train_loader, test_loader, device, epochs=10  # Changed to 10 epochs
)

# Save the original model's state dictionary
torch.save(original_model.state_dict(), 'expanded_cnn_original.pth')
print("Original model saved to expanded_cnn_original.pth")

# ============================
# 7. Define Structured Pruning Functions
# ============================

def apply_structured_pruning(model, amount=0.2):
    """
    Applies structured pruning to all Conv2d layers in the model.

    Args:
        model (nn.Module): The neural network model.
        amount (float): The fraction of output channels to prune in each Conv layer (e.g., 0.2 for 20%).

    Returns:
        None
    """
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d):
            # Prune entire output channels (filters) based on L1 norm
            prune.ln_structured(module, name='weight', amount=amount, n=1, dim=0)
            print(f"Applied {amount*100}% structured pruning to Conv layer: {name}")

def remove_pruning_reparametrization(model):
    """
    Removes pruning reparameterization from all Conv2d layers in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        None
    """
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d):
            prune.remove(module, 'weight')
            print(f"Removed pruning reparameterization from layer: {name}")

# ============================
# 8. Define Function to Capture Activation Maps
# ============================

def get_activation(model, layer_name, input_image):
    """
    Captures the output of a specified layer using forward hooks.

    Args:
        model (nn.Module): The neural network model.
        layer_name (str): The name of the layer to capture.
        input_image (torch.Tensor): The input image tensor.

    Returns:
        activation (torch.Tensor): The activation output from the specified layer.
    """
    activation = {}

    def hook_fn(module, input, output):
        activation['output'] = output.detach()

    # Register hook
    hook = None
    for name, module in model.named_modules():
        if name == layer_name:
            hook = module.register_forward_hook(hook_fn)
            break

    if hook is None:
        print(f"Layer {layer_name} not found in the model.")
        return None

    # Forward pass
    model.eval()
    with torch.no_grad():
        _ = model(input_image.to(device))

    # Remove hook
    hook.remove()

    return activation.get('output')

def display_activation_maps_ranked(model, layer_name, input_image, top_k=32):
    """
    Displays activation maps of a specified layer ranked by the filter's L1 norm.

    Args:
        model (nn.Module): The neural network model.
        layer_name (str): The name of the layer to capture.
        input_image (torch.Tensor): The input image tensor.
        top_k (int): Number of top activation maps to display.

    Returns:
        None
    """
    # Get activation from the specified layer
    activation = get_activation(model, layer_name, input_image)

    if activation is None:
        return

    # Convert activation to CPU and NumPy
    activation_np = activation.cpu().numpy()  # Shape: (1, C, H, W)

    # Compute L1 norm for each channel
    l1_norm = np.sum(np.abs(activation_np), axis=(0, 2, 3))  # Shape: (C,)

    # Rank channels based on L1 norm
    sorted_indices = np.argsort(l1_norm)[::-1]  # Descending order

    # Select top_k channels
    top_k = min(top_k, activation_np.shape[1])
    top_indices = sorted_indices[:top_k]
    top_activation = activation_np[0, top_indices, :, :]  # Shape: (top_k, H, W)

    # Plot the top_k activation maps
    num_cols = 8
    num_rows = top_k // num_cols + int(top_k % num_cols != 0)

    plt.figure(figsize=(num_cols * 2, num_rows * 2))
    plt.suptitle(f"Top {top_k} Activation Maps After {layer_name} Ranked by L1 Norm")

    for i in range(top_k):
        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(top_activation[i], cmap='gray')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# ============================
# 9. Define Function to Count Trainable Parameters
# ============================

def count_trainable_parameters(model):
    """
    Counts the total number of trainable parameters in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        count (int): Total number of trainable parameters.
    """
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ============================
# 10. Define Function to Calculate Pruned Parameters
# ============================

def calculate_pruned_parameters(model):
    """
    Calculates the total number of parameters and the number of pruned (zeroed) parameters in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        total_params (int): Total number of parameters.
        pruned_params (int): Number of pruned (zeroed) parameters.
    """
    total_params = 0
    pruned_params = 0
    for name, param in model.named_parameters():
        if 'weight' in name:
            total_params += param.numel()
            pruned_params += torch.sum(param == 0).item()
    return total_params, pruned_params

# ============================
# 11. Apply Structured Pruning at Multiple Levels and Measure Performance
# ============================

# Define pruning percentages (80%, 90%, 95%, 99%)
pruning_percentages = [0.8, 0.9, 0.95, 0.99]  # Added 80% and 99%

# Initialize a list to store results
results = []

# ============================
# 11.1. Display Activation Maps of Original Model Ranked by conv1 Filters' L1 Norm
# ============================

print("\n=== Displaying conv1 Activation Maps of Original Model Ranked by L1 Norm ===")
display_activation_maps_ranked(original_model, 'conv1', single_img.unsqueeze(0), top_k=32)  # Adjust top_k as needed

# ============================
# 11.2. Apply Pruning, Redefine Model, Train, and Record Results
# ============================

for pct in pruning_percentages:
    print(f"\n=== Applying {int(pct*100)}% Structured Pruning ===")

    # Load the original model
    pruned_model = ExpandedCNN().to(device)
    pruned_model.load_state_dict(torch.load('expanded_cnn_original.pth', map_location=device))
    pruned_model.eval()

    # Apply structured pruning
    apply_structured_pruning(pruned_model, amount=pct)

    # Remove pruning reparametrization to finalize the model
    remove_pruning_reparametrization(pruned_model)

    # ============================
    # 11.2.1. Identify Kept Output Channels for Each Conv Layer
    # ============================

    def get_keep_indices(model):
        keep_indices = {}
        for name, module in model.named_modules():
            if isinstance(module, nn.Conv2d):
                # Calculate L1 norm across (in_channels, H, W) for each output channel
                l1_norm = module.weight.data.abs().sum(dim=(1,2,3))
                # Get indices of channels that were not pruned (i.e., L1 norm > 0)
                indices = torch.nonzero(l1_norm > 0).squeeze().cpu().numpy()
                if indices.size == 0:
                    print(f"All channels have been pruned in layer: {name}. Skipping this pruning percentage.")
                    return None
                if indices.ndim == 0:
                    indices = np.array([indices])
                keep_indices[name] = indices
        return keep_indices

    keep_indices = get_keep_indices(pruned_model)
    if keep_indices is None:
        continue  # Skip to next pruning percentage if any Conv layer is fully pruned

    # ============================
    # 11.2.2. Redefine the Model with Reduced Channels
    # ============================

    class PrunedExpandedCNN_AllLayers(nn.Module):
        def __init__(self, keep_conv1_out_channels, keep_conv2_out_channels, keep_conv3_out_channels, keep_conv4_out_channels):
            super(PrunedExpandedCNN_AllLayers, self).__init__()
            # First block with reduced conv1 and conv2
            self.conv1 = nn.Conv2d(1, keep_conv1_out_channels, kernel_size=3, padding=1)
            self.relu1 = nn.ReLU(inplace=True)
            self.conv2 = nn.Conv2d(keep_conv1_out_channels, keep_conv2_out_channels, kernel_size=3, padding=1)
            self.relu2 = nn.ReLU(inplace=True)
            self.pool1 = nn.MaxPool2d(2, 2)
            self.dropout1 = nn.Dropout(0.25)

            # Second block with reduced conv3 and conv4
            self.conv3 = nn.Conv2d(keep_conv2_out_channels, keep_conv3_out_channels, kernel_size=3, padding=1)
            self.relu3 = nn.ReLU(inplace=True)
            self.conv4 = nn.Conv2d(keep_conv3_out_channels, keep_conv4_out_channels, kernel_size=3, padding=1)
            self.relu4 = nn.ReLU(inplace=True)
            self.pool2 = nn.MaxPool2d(2, 2)
            self.dropout2 = nn.Dropout(0.25)

            # Fully connected layers
            self.fc1 = nn.Linear(keep_conv4_out_channels * 7 * 7, 512)
            self.relu5 = nn.ReLU(inplace=True)
            self.dropout3 = nn.Dropout(0.5)
            self.fc2 = nn.Linear(512, 10)

        def forward(self, x):
            # First block
            x = self.conv1(x)
            x = self.relu1(x)
            x = self.conv2(x)
            x = self.relu2(x)
            x = self.pool1(x)
            x = self.dropout1(x)

            # Second block
            x = self.conv3(x)
            x = self.relu3(x)
            x = self.conv4(x)
            x = self.relu4(x)
            x = self.pool2(x)
            x = self.dropout2(x)

            # Fully connected layers
            x = x.view(x.size(0), -1)  # Flatten
            x = self.fc1(x)
            x = self.relu5(x)
            x = self.dropout3(x)
            x = self.fc2(x)
            return F.log_softmax(x, dim=1)

    # Extract keep indices for each conv layer
    keep_conv1 = keep_indices.get('conv1', None)
    keep_conv2 = keep_indices.get('conv2', None)
    keep_conv3 = keep_indices.get('conv3', None)
    keep_conv4 = keep_indices.get('conv4', None)

    # If any layer has no keep indices, skip pruning
    if keep_conv1 is None or keep_conv2 is None or keep_conv3 is None or keep_conv4 is None:
        print("Skipping pruning due to complete pruning in one of the Conv layers.")
        continue

    # Instantiate the pruned and redefined model with correct channel counts
    pruned_redefined_model = PrunedExpandedCNN_AllLayers(
        keep_conv1_out_channels=len(keep_conv1),
        keep_conv2_out_channels=len(keep_conv2),
        keep_conv3_out_channels=len(keep_conv3),
        keep_conv4_out_channels=len(keep_conv4)
    ).to(device)

    print("\nRedefining the pruned model with reduced channels:")
    print(pruned_redefined_model)

    # ============================
    # 11.2.3. Transfer Weights Correctly Without Slicing Input Channels
    # ============================

    # Transfer conv1 weights and biases
    pruned_redefined_model.conv1.weight.data = pruned_model.conv1.weight.data[keep_conv1].clone()
    pruned_redefined_model.conv1.bias.data = pruned_model.conv1.bias.data[keep_conv1].clone()
    print(f"Transferred conv1 weights and biases. New conv1 out_channels: {len(keep_conv1)}")

    # Transfer conv2 weights and biases
    # conv2.weight shape: [original_out_channels, original_in_channels, 3, 3]
    # After pruning:
    # pruned_redefined_model.conv2 expects [keep_conv2_out_channels, keep_conv1_out_channels, 3, 3]
    pruned_redefined_model.conv2.weight.data = pruned_model.conv2.weight.data[keep_conv2][:, keep_conv1].clone()
    pruned_redefined_model.conv2.bias.data = pruned_model.conv2.bias.data[keep_conv2].clone()
    print(f"Transferred conv2 weights and biases. New conv2 out_channels: {len(keep_conv2)}")

    # Transfer conv3 weights and biases
    # pruned_redefined_model.conv3 expects [keep_conv3_out_channels, keep_conv2_out_channels, 3, 3]
    pruned_redefined_model.conv3.weight.data = pruned_model.conv3.weight.data[keep_conv3][:, keep_conv2].clone()
    pruned_redefined_model.conv3.bias.data = pruned_model.conv3.bias.data[keep_conv3].clone()
    print(f"Transferred conv3 weights and biases. New conv3 out_channels: {len(keep_conv3)}")

    # Transfer conv4 weights and biases
    # pruned_redefined_model.conv4 expects [keep_conv4_out_channels, keep_conv3_out_channels, 3, 3]
    pruned_redefined_model.conv4.weight.data = pruned_model.conv4.weight.data[keep_conv4][:, keep_conv3].clone()
    pruned_redefined_model.conv4.bias.data = pruned_model.conv4.bias.data[keep_conv4].clone()
    print(f"Transferred conv4 weights and biases. New conv4 out_channels: {len(keep_conv4)}")

    # ============================
    # **Important Correction Here**:
    # Do **NOT** transfer fc1's weights and biases from the original model
    # because its input dimensions have changed.
    # Instead, initialize fc1 in the pruned model.
    # ============================

    # Initialize fc1 weights and biases
    nn.init.kaiming_normal_(pruned_redefined_model.fc1.weight, nonlinearity='relu')
    if pruned_redefined_model.fc1.bias is not None:
        nn.init.constant_(pruned_redefined_model.fc1.bias, 0)
    print(f"Initialized fc1 weights and biases with proper dimensions.")

    # Transfer fc2 weights and biases directly as their dimensions remain unchanged
    pruned_redefined_model.fc2.weight.data = pruned_model.fc2.weight.data.clone()
    pruned_redefined_model.fc2.bias.data = pruned_model.fc2.bias.data.clone()
    print(f"Transferred fc2 weights and biases.")

    print("Successfully redefined the model with reduced channels.")

    # ============================
    # 11.2.4. Train and Evaluate the Pruned and Re-defined Model
    # ============================

    print(f"\nTraining Pruned and Re-defined Model with {int(pct*100)}% Pruning:")
    test_acc, train_time = train_and_evaluate(pruned_redefined_model, train_loader, test_loader, device, epochs=10)  # Changed to 10 epochs

    # Count the number of trainable parameters
    num_params = count_trainable_parameters(pruned_redefined_model)

    # Calculate total and pruned parameters (pruned_params will be 0 as pruned channels are removed)
    total_params, pruned_params = calculate_pruned_parameters(pruned_redefined_model)

    # Append the results without "Total Parameters Pruned"
    results.append({
        'Pruning Percentage': f"{int(pct*100)}%",
        'Trainable Parameters': f"{num_params}",
        'Training Time (s)': f"{train_time:.2f}",
        'Test Accuracy': f"{test_acc*100:.2f}%"
    })

    # ============================
    # 11.2.5. Display conv1 Activation Maps for the Pruned Model
    # ============================

    print(f"\n=== Displaying conv1 Activation Maps for pruned model with {int(pct*100)}% pruning ===")
    display_activation_maps_ranked(pruned_redefined_model, 'conv1', single_img.unsqueeze(0), top_k=len(keep_conv1))

    # Save the pruned and redefined model after fine-tuning
    model_save_path = f"expanded_cnn_pruned_{int(pct*100)}.pth"
    torch.save(pruned_redefined_model.state_dict(), model_save_path)
    print(f"Pruned and Re-defined model saved to {model_save_path}")

# ============================
# 12. Evaluate and Record Original Model Performance Again
# ============================

print("\n=== Training Original Model for Comparison ===")
original_test_acc_new, original_train_time_new = train_and_evaluate(original_model, train_loader, test_loader, device, epochs=10)  # Changed to 10 epochs

# Count the number of trainable parameters for the original model
original_num_params = count_trainable_parameters(original_model)

# Append the original model's results at the beginning (without "Total Parameters Pruned")
results.insert(0, {
    'Pruning Percentage': "0%",
    'Trainable Parameters': f"{original_num_params}",
    'Training Time (s)': f"{original_train_time_new:.2f}",
    'Test Accuracy': f"{original_test_acc_new*100:.2f}%"
})

# ============================
# 13. Create and Display Results Table
# ============================

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Reorder columns for better readability
results_df = results_df[['Pruning Percentage', 'Trainable Parameters', 'Training Time (s)', 'Test Accuracy']]

# Display the table
print("\n=== Training Time, Test Accuracy, and Trainable Parameters ===")
print(results_df.to_string(index=False))
```

```python
# ============================
# 1. Import Necessary Libraries
# ============================

# Install torch and torchvision if not already installed
# (Uncomment the following line if running in an environment where torch isn't installed)
# !pip install torch torchvision --quiet

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.utils.prune as prune
import time
import pandas as pd

# ============================
# 2. Set Device to CPU or CUDA
# ============================

# Set device to CUDA if available, else CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ============================
# 3. Define Data Transformations and Load Datasets
# ============================

# Define transformations: convert images to tensors and normalize them
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # MNIST mean and std
])

# Define batch size
batch_size = 64

# Download and load training and test datasets
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Define DataLoaders with drop_last=True to ensure all batches are the same size
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, drop_last=True)

print(f"Number of training batches: {len(train_loader)}")
print(f"Number of testing batches: {len(test_loader)}")

# ============================
# 4. Display a Single Training Example
# ============================

# Function to unnormalize and convert tensor to numpy array for display
def imshow(img, title):
    img = img.numpy().transpose((1, 2, 0))  # Convert from Tensor image
    mean = np.array([0.1307])
    std = np.array([0.3081])
    img = std * img + mean  # Unnormalize
    img = np.clip(img, 0, 1)  # Clip to valid range
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Get one batch of training data
dataiter = iter(train_loader)
images, labels = next(dataiter)

# Select the first image and label from the batch
single_img = images[0]
single_label = labels[0]

# Display the single training example
print("Displaying a single training example:")
imshow(single_img.cpu(), f"Label: {single_label.item()}")

# ============================
# 5. Define the Expanded CNN Model
# ============================

class ExpandedCNN(nn.Module):
    def __init__(self, conv1_out_channels=32, conv2_out_channels=64, conv3_out_channels=128, conv4_out_channels=256):
        super(ExpandedCNN, self).__init__()
        # First block
        self.conv1 = nn.Conv2d(1, conv1_out_channels, kernel_size=3, padding=1)  # 1x28x28 -> conv1_out_channels x28x28
        self.relu1 = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(conv1_out_channels, conv2_out_channels, kernel_size=3, padding=1)  # conv1_out_channels x28x28 -> conv2_out_channels x28x28
        self.relu2 = nn.ReLU(inplace=True)
        self.pool1 = nn.MaxPool2d(2, 2)  # conv2_out_channels x28x28 -> conv2_out_channels x14x14
        self.dropout1 = nn.Dropout(0.25)

        # Second block
        self.conv3 = nn.Conv2d(conv2_out_channels, conv3_out_channels, kernel_size=3, padding=1)  # conv2_out_channels x14x14 -> conv3_out_channels x14x14
        self.relu3 = nn.ReLU(inplace=True)
        self.conv4 = nn.Conv2d(conv3_out_channels, conv4_out_channels, kernel_size=3, padding=1)  # conv3_out_channels x14x14 -> conv4_out_channels x14x14
        self.relu4 = nn.ReLU(inplace=True)
        self.pool2 = nn.MaxPool2d(2, 2)  # conv4_out_channels x14x14 -> conv4_out_channels x7x7
        self.dropout2 = nn.Dropout(0.25)

        # Fully connected layers
        self.fc1 = nn.Linear(conv4_out_channels * 7 * 7, 512)
        self.relu5 = nn.ReLU(inplace=True)
        self.dropout3 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 10)  # 10 output classes for MNIST

    def forward(self, x):
        # First block
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool1(x)
        x = self.dropout1(x)

        # Second block
        x = self.conv3(x)
        x = self.relu3(x)
        x = self.conv4(x)
        x = self.relu4(x)
        x = self.pool2(x)
        x = self.dropout2(x)

        # Fully connected layers
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc1(x)
        x = self.relu5(x)
        x = self.dropout3(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)  # Log-Softmax for classification

# ============================
# 6. Initialize and Train the Original Model
# ============================

def train_and_evaluate(model, train_loader, test_loader, device, epochs=10):
    """
    Trains the model for a specified number of epochs and evaluates on the test set.

    Args:
        model (nn.Module): The neural network model.
        train_loader (DataLoader): DataLoader for the training dataset.
        test_loader (DataLoader): DataLoader for the test dataset.
        device (torch.device): Device to perform computations on.
        epochs (int): Number of training epochs.

    Returns:
        test_accuracy (float): Accuracy of the model on the test set after training.
        training_time (float): Time taken to train the model in seconds.
    """
    criterion = nn.NLLLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    start_time = time.time()

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()  # Zero the parameter gradients
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = correct / total

        # Evaluate on test data
        model.eval()
        test_loss = 0.0
        correct_test = 0
        total_test = 0

        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                test_loss += loss.item() * images.size(0)
                _, predicted = torch.max(outputs.data, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        epoch_test_loss = test_loss / len(test_loader.dataset)
        epoch_test_acc = correct_test / total_test

        print(f"Epoch {epoch}/{epochs} | "
              f"Train Loss: {epoch_loss:.4f} | Train Acc: {epoch_acc:.4f} | "
              f"Test Loss: {epoch_test_loss:.4f} | Test Acc: {epoch_test_acc:.4f}")

    end_time = time.time()
    training_time = end_time - start_time

    return epoch_test_acc, training_time

# Initialize the original model
original_model = ExpandedCNN().to(device)
print("\nOriginal Model Architecture:")
print(original_model)

# Train and evaluate the original model
print("\nTraining Original Model:")
original_test_acc, original_train_time = train_and_evaluate(
    original_model, train_loader, test_loader, device, epochs=10  # Changed to 10 epochs
)

# Save the original model's state dictionary
torch.save(original_model.state_dict(), 'expanded_cnn_original.pth')
print("Original model saved to expanded_cnn_original.pth")

# ============================
# 7. Define Structured Pruning Functions
# ============================

def apply_structured_pruning(model, amount=0.2):
    """
    Applies structured pruning to all Conv2d and Linear layers in the model.

    Args:
        model (nn.Module): The neural network model.
        amount (float): The fraction of output channels or neurons to prune in each layer (e.g., 0.2 for 20%).

    Returns:
        None
    """
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d) or isinstance(module, nn.Linear):
            # Prune entire output channels (filters) or neurons based on L1 norm
            prune.ln_structured(module, name='weight', amount=amount, n=1, dim=0)
            layer_type = 'Conv layer' if isinstance(module, nn.Conv2d) else 'Linear layer'
            print(f"Applied {amount*100}% structured pruning to {layer_type}: {name}")

def remove_pruning_reparametrization(model):
    """
    Removes pruning reparameterization from all Conv2d and Linear layers in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        None
    """
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d) or isinstance(module, nn.Linear):
            prune.remove(module, 'weight')
            layer_type = 'Conv layer' if isinstance(module, nn.Conv2d) else 'Linear layer'
            print(f"Removed pruning reparameterization from {layer_type}: {name}")

# ============================
# 8. Define Function to Capture Activation Maps
# ============================

def get_activation(model, layer_name, input_image):
    """
    Captures the output of a specified layer using forward hooks.

    Args:
        model (nn.Module): The neural network model.
        layer_name (str): The name of the layer to capture.
        input_image (torch.Tensor): The input image tensor.

    Returns:
        activation (torch.Tensor): The activation output from the specified layer.
    """
    activation = {}

    def hook_fn(module, input, output):
        activation['output'] = output.detach()

    # Register hook
    hook = None
    for name, module in model.named_modules():
        if name == layer_name:
            hook = module.register_forward_hook(hook_fn)
            break

    if hook is None:
        print(f"Layer {layer_name} not found in the model.")
        return None

    # Forward pass
    model.eval()
    with torch.no_grad():
        _ = model(input_image.to(device))

    # Remove hook
    hook.remove()

    return activation.get('output')

def display_activation_maps_ranked(model, layer_name, input_image, top_k=32):
    """
    Displays activation maps of a specified layer ranked by the filter's L1 norm.

    Args:
        model (nn.Module): The neural network model.
        layer_name (str): The name of the layer to capture.
        input_image (torch.Tensor): The input image tensor.
        top_k (int): Number of top activation maps to display.

    Returns:
        None
    """
    # Get activation from the specified layer
    activation = get_activation(model, layer_name, input_image)

    if activation is None:
        return

    # Convert activation to CPU and NumPy
    activation_np = activation.cpu().numpy()  # Shape: (1, C, H, W)

    # Compute L1 norm for each channel
    l1_norm = np.sum(np.abs(activation_np), axis=(0, 2, 3))  # Shape: (C,)

    # Rank channels based on L1 norm
    sorted_indices = np.argsort(l1_norm)[::-1]  # Descending order

    # Select top_k channels
    top_k = min(top_k, activation_np.shape[1])
    top_indices = sorted_indices[:top_k]
    top_activation = activation_np[0, top_indices, :, :]  # Shape: (top_k, H, W)

    # Plot the top_k activation maps
    num_cols = 8
    num_rows = top_k // num_cols + int(top_k % num_cols != 0)

    plt.figure(figsize=(num_cols * 2, num_rows * 2))
    plt.suptitle(f"Top {top_k} Activation Maps After {layer_name} Ranked by L1 Norm")

    for i in range(top_k):
        plt.subplot(num_rows, num_cols, i + 1)
        plt.imshow(top_activation[i], cmap='gray')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# ============================
# 9. Define Function to Count Trainable Parameters
# ============================

def count_trainable_parameters(model):
    """
    Counts the total number of trainable parameters in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        count (int): Total number of trainable parameters.
    """
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ============================
# 10. Define Function to Calculate Pruned Parameters
# ============================

def calculate_pruned_parameters(model):
    """
    Calculates the total number of parameters and the number of pruned (zeroed) parameters in the model.

    Args:
        model (nn.Module): The neural network model.

    Returns:
        total_params (int): Total number of parameters.
        pruned_params (int): Number of pruned (zeroed) parameters.
    """
    total_params = 0
    pruned_params = 0
    for name, param in model.named_parameters():
        if 'weight' in name or 'bias' in name:
            total_params += param.numel()
            pruned_params += torch.sum(param == 0).item()
    return total_params, pruned_params

# ============================
# 11. Apply Structured Pruning at Multiple Levels and Measure Performance
# ============================

# Define pruning percentages (80%, 90%, 95%, 99%)
pruning_percentages = [0.8, 0.9, 0.95, 0.99]  # Added 80% and 99%

# Initialize a list to store results
results = []

# ============================
# 11.1. Display Activation Maps of Original Model Ranked by conv1 Filters' L1 Norm
# ============================

print("\n=== Displaying conv1 Activation Maps of Original Model Ranked by L1 Norm ===")
display_activation_maps_ranked(original_model, 'conv1', single_img.unsqueeze(0), top_k=32)  # Adjust top_k as needed

# ============================
# 11.2. Apply Pruning, Redefine Model, Train, and Record Results
# ============================

for pct in pruning_percentages:
    print(f"\n=== Applying {int(pct*100)}% Structured Pruning ===")

    # Load the original model
    pruned_model = ExpandedCNN().to(device)
    pruned_model.load_state_dict(torch.load('expanded_cnn_original.pth', map_location=device))
    pruned_model.eval()

    # Apply structured pruning to both Conv2d and Linear layers
    apply_structured_pruning(pruned_model, amount=pct)

    # Remove pruning reparametrization to finalize the model
    remove_pruning_reparametrization(pruned_model)

    # ============================
    # 11.2.1. Identify Kept Output Channels and Neurons for Each Layer
    # ============================

    def get_keep_indices_all_layers(model):
        keep_indices = {}
        for name, module in model.named_modules():
            if isinstance(module, nn.Conv2d):
                # For Conv2d: Prune output channels based on L1 norm
                l1_norm = module.weight.data.abs().sum(dim=(1,2,3))
                indices = torch.nonzero(l1_norm > 0).squeeze().cpu().numpy()
                if indices.size == 0:
                    print(f"All channels have been pruned in Conv layer: {name}. Skipping this pruning percentage.")
                    return None
                if indices.ndim == 0:
                    indices = np.array([indices])
                keep_indices[name] = indices
            elif isinstance(module, nn.Linear):
                # For Linear: Prune output neurons based on L1 norm
                l1_norm = module.weight.data.abs().sum(dim=1)
                indices = torch.nonzero(l1_norm > 0).squeeze().cpu().numpy()
                if indices.size == 0:
                    print(f"All neurons have been pruned in Linear layer: {name}. Skipping this pruning percentage.")
                    return None
                if indices.ndim == 0:
                    indices = np.array([indices])
                keep_indices[name] = indices
        return keep_indices

    keep_indices = get_keep_indices_all_layers(pruned_model)
    if keep_indices is None:
        continue  # Skip to next pruning percentage if any layer is fully pruned

    # ============================
    # 11.2.2. Redefine the Model with Reduced Channels and Neurons
    # ============================

    class PrunedExpandedCNN_AllLayers(nn.Module):
        def __init__(self, keep_conv1_out_channels, keep_conv2_out_channels, keep_conv3_out_channels, keep_conv4_out_channels,
                     keep_fc1_out_features, keep_fc2_out_features=10):
            super(PrunedExpandedCNN_AllLayers, self).__init__()
            # First block with reduced conv1 and conv2
            self.conv1 = nn.Conv2d(1, keep_conv1_out_channels, kernel_size=3, padding=1)
            self.relu1 = nn.ReLU(inplace=True)
            self.conv2 = nn.Conv2d(keep_conv1_out_channels, keep_conv2_out_channels, kernel_size=3, padding=1)
            self.relu2 = nn.ReLU(inplace=True)
            self.pool1 = nn.MaxPool2d(2, 2)
            self.dropout1 = nn.Dropout(0.25)

            # Second block with reduced conv3 and conv4
            self.conv3 = nn.Conv2d(keep_conv2_out_channels, keep_conv3_out_channels, kernel_size=3, padding=1)
            self.relu3 = nn.ReLU(inplace=True)
            self.conv4 = nn.Conv2d(keep_conv3_out_channels, keep_conv4_out_channels, kernel_size=3, padding=1)
            self.relu4 = nn.ReLU(inplace=True)
            self.pool2 = nn.MaxPool2d(2, 2)
            self.dropout2 = nn.Dropout(0.25)

            # Fully connected layers with reduced neurons
            self.fc1 = nn.Linear(keep_conv4_out_channels * 7 * 7, keep_fc1_out_features)
            self.relu5 = nn.ReLU(inplace=True)
            self.dropout3 = nn.Dropout(0.5)
            self.fc2 = nn.Linear(keep_fc1_out_features, keep_fc2_out_features)

        def forward(self, x):
            # First block
            x = self.conv1(x)
            x = self.relu1(x)
            x = self.conv2(x)
            x = self.relu2(x)
            x = self.pool1(x)
            x = self.dropout1(x)

            # Second block
            x = self.conv3(x)
            x = self.relu3(x)
            x = self.conv4(x)
            x = self.relu4(x)
            x = self.pool2(x)
            x = self.dropout2(x)

            # Fully connected layers
            x = x.view(x.size(0), -1)  # Flatten
            x = self.fc1(x)
            x = self.relu5(x)
            x = self.dropout3(x)
            x = self.fc2(x)
            return F.log_softmax(x, dim=1)

    # Extract keep indices for each layer
    keep_conv1 = keep_indices.get('conv1', None)
    keep_conv2 = keep_indices.get('conv2', None)
    keep_conv3 = keep_indices.get('conv3', None)
    keep_conv4 = keep_indices.get('conv4', None)
    keep_fc1 = keep_indices.get('fc1', None)
    keep_fc2 = keep_indices.get('fc2', None)  # Should remain 10

    # Validate all necessary layers have kept indices
    if keep_conv1 is None or keep_conv2 is None or keep_conv3 is None or keep_conv4 is None or keep_fc1 is None:
        print("Skipping pruning due to complete pruning in one of the layers.")
        continue

    # Instantiate the pruned and redefined model with correct channel counts and neurons
    pruned_redefined_model = PrunedExpandedCNN_AllLayers(
        keep_conv1_out_channels=len(keep_conv1),
        keep_conv2_out_channels=len(keep_conv2),
        keep_conv3_out_channels=len(keep_conv3),
        keep_conv4_out_channels=len(keep_conv4),
        keep_fc1_out_features=len(keep_fc1),
        keep_fc2_out_features=10  # Output classes remain the same
    ).to(device)

    print("\nRedefining the pruned model with reduced channels and neurons:")
    print(pruned_redefined_model)

    # ============================
    # 11.2.3. Transfer Weights Correctly Without Slicing Input Channels or Neurons
    # ============================

    # Function to compute input feature indices for fc1 based on kept conv4 channels
    def get_fc1_in_features_indices(keep_conv4, spatial_dim=7*7):
        """
        Computes the list of input feature indices for fc1 based on kept conv4 channels.

        Args:
            keep_conv4 (list or np.array): Indices of kept conv4 output channels.
            spatial_dim (int): Spatial dimension size (height * width). Default is 7*7 for MNIST.

        Returns:
            keep_fc1_in_features (list): List of input feature indices to keep for fc1.
        """
        keep_fc1_in_features = []
        for c in keep_conv4:
            start_idx = c * spatial_dim
            end_idx = (c + 1) * spatial_dim
            keep_fc1_in_features.extend(range(start_idx, end_idx))
        return keep_fc1_in_features

    # Compute input feature indices for fc1
    keep_fc1_in_features = get_fc1_in_features_indices(keep_conv4, spatial_dim=7*7)

    # Verify that the number of input features matches
    expected_fc1_in_features = pruned_redefined_model.fc1.in_features
    actual_fc1_in_features = len(keep_fc1_in_features)
    print(f"Expected fc1 in_features: {expected_fc1_in_features}, Actual kept input features: {actual_fc1_in_features}")

    if expected_fc1_in_features != actual_fc1_in_features:
        print("Mismatch in fc1 input features after pruning. Adjusting the model accordingly.")
        # Optionally, adjust the fc1 layer's in_features or handle the mismatch
        # For simplicity, we'll assume the calculation is correct

    # Transfer conv1 weights and biases
    pruned_redefined_model.conv1.weight.data = pruned_model.conv1.weight.data[keep_conv1, :, :, :].clone()
    pruned_redefined_model.conv1.bias.data = pruned_model.conv1.bias.data[keep_conv1].clone()
    print(f"Transferred conv1 weights and biases. New conv1 out_channels: {len(keep_conv1)}")

    # Transfer conv2 weights and biases
    # pruned_redefined_model.conv2 expects [keep_conv2_out_channels, keep_conv1_out_channels, 3, 3]
    pruned_redefined_model.conv2.weight.data = pruned_model.conv2.weight.data[keep_conv2, :, :, :].clone()[:, keep_conv1, :, :].clone()
    pruned_redefined_model.conv2.bias.data = pruned_model.conv2.bias.data[keep_conv2].clone()
    print(f"Transferred conv2 weights and biases. New conv2 out_channels: {len(keep_conv2)}")

    # Transfer conv3 weights and biases
    # pruned_redefined_model.conv3 expects [keep_conv3_out_channels, keep_conv2_out_channels, 3, 3]
    pruned_redefined_model.conv3.weight.data = pruned_model.conv3.weight.data[keep_conv3, :, :, :].clone()[:, keep_conv2, :, :].clone()
    pruned_redefined_model.conv3.bias.data = pruned_model.conv3.bias.data[keep_conv3].clone()
    print(f"Transferred conv3 weights and biases. New conv3 out_channels: {len(keep_conv3)}")

    # Transfer conv4 weights and biases
    # pruned_redefined_model.conv4 expects [keep_conv4_out_channels, keep_conv3_out_channels, 3, 3]
    pruned_redefined_model.conv4.weight.data = pruned_model.conv4.weight.data[keep_conv4, :, :, :].clone()[:, keep_conv3, :, :].clone()
    pruned_redefined_model.conv4.bias.data = pruned_model.conv4.bias.data[keep_conv4].clone()
    print(f"Transferred conv4 weights and biases. New conv4 out_channels: {len(keep_conv4)}")

    # Transfer fc1 weights and biases (only the kept output neurons and kept input features)
    pruned_redefined_model.fc1.weight.data = pruned_model.fc1.weight.data[keep_fc1, :].clone()[:, keep_fc1_in_features]
    pruned_redefined_model.fc1.bias.data = pruned_model.fc1.bias.data[keep_fc1].clone()
    print(f"Transferred fc1 weights and biases. New fc1 out_features: {len(keep_fc1)}")

    # Transfer fc2 weights and biases (adjust input features based on fc1 pruning)
    # pruned_redefined_model.fc2 expects [keep_fc2_out_features, keep_fc1_out_features]
    pruned_redefined_model.fc2.weight.data = pruned_model.fc2.weight.data[:, keep_fc1].clone()
    pruned_redefined_model.fc2.bias.data = pruned_model.fc2.bias.data.clone()
    print(f"Transferred fc2 weights and biases. New fc2 in_features: {len(keep_fc1)}")

    print("Successfully redefined the model with reduced channels and neurons.")

    # ============================
    # 11.2.4. Train and Evaluate the Pruned and Re-defined Model
    # ============================

    print(f"\nTraining Pruned and Re-defined Model with {int(pct*100)}% Pruning:")
    test_acc, train_time = train_and_evaluate(pruned_redefined_model, train_loader, test_loader, device, epochs=10)  # Changed to 10 epochs

    # Count the number of trainable parameters
    num_params = count_trainable_parameters(pruned_redefined_model)

    # Calculate total and pruned parameters (pruned_params will be 0 as pruned channels and neurons are removed)
    total_params, pruned_params = calculate_pruned_parameters(pruned_redefined_model)

    # Append the results without "Total Parameters Pruned"
    results.append({
        'Pruning Percentage': f"{int(pct*100)}%",
        'Trainable Parameters': f"{num_params}",
        'Training Time (s)': f"{train_time:.2f}",
        'Test Accuracy': f"{test_acc*100:.2f}%"
    })

    # ============================
    # 11.2.5. Display conv1 Activation Maps for the Pruned Model
    # ============================

    print(f"\n=== Displaying conv1 Activation Maps for pruned model with {int(pct*100)}% pruning ===")
    display_activation_maps_ranked(pruned_redefined_model, 'conv1', single_img.unsqueeze(0), top_k=len(keep_conv1))

    # Save the pruned and redefined model after fine-tuning
    model_save_path = f"expanded_cnn_pruned_{int(pct*100)}.pth"
    torch.save(pruned_redefined_model.state_dict(), model_save_path)
    print(f"Pruned and Re-defined model saved to {model_save_path}")

# ============================
# 12. Evaluate and Record Original Model Performance Again
# ============================

print("\n=== Training Original Model for Comparison ===")
original_test_acc_new, original_train_time_new = train_and_evaluate(original_model, train_loader, test_loader, device, epochs=10)  # Changed to 10 epochs

# Count the number of trainable parameters for the original model
original_num_params = count_trainable_parameters(original_model)

# Append the original model's results at the beginning (without "Total Parameters Pruned")
results.insert(0, {
    'Pruning Percentage': "0%",
    'Trainable Parameters': f"{original_num_params}",
    'Training Time (s)': f"{original_train_time_new:.2f}",
    'Test Accuracy': f"{original_test_acc_new*100:.2f}%"
})

# ============================
# 13. Create and Display Results Table
# ============================

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Reorder columns for better readability
results_df = results_df[['Pruning Percentage', 'Trainable Parameters', 'Training Time (s)', 'Test Accuracy']]

# Display the table
print("\n=== Training Time, Test Accuracy, and Trainable Parameters ===")
print(results_df.to_string(index=False))
```
